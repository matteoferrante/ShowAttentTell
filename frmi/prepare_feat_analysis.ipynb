```python
import  os
import glob
from os.path import join as opj
import pandas as pd
import numpy as np
```

```python
sub="sub-CSI1"
data_path=f"/space/matteo/BOLD5000/ds001499/{sub}"
```

```python
ses="ses-01"

ses_path=os.path.join(data_path,ses,"func")
func_files=[i for i in os.listdir(ses_path) if "_bold.nii.gz" in i]

func_files=sorted(func_files)

events_files=[i.replace("_bold.nii.gz","_events.tsv") for i in func_files]
```

```python
func_files
```

```python
index=0


print(f"[FUNC]\n\n{os.path.join(ses_path,func_files[index])}")
df=pd.read_csv(os.path.join(ses_path,events_files[0]),sep="\t")

df_three=pd.DataFrame(columns=["onset","duration","magnitude"])
df_three.onset=df.onset
df_three.duration=df.duration
df_three.magnitude=np.ones(len(df))

df_three_coco=df_three[df.ImgType=="coco"]
df_three_imagenet=df_three[df.ImgType=="imagenet"]
df_three_scene=df_three[df.ImgType=="scenes"]

os.makedirs(f"three_columns_{sub}_{ses}",exist_ok=True)

df_three_coco.to_csv(f"three_columns_{sub}_{ses}/custom_three_coco_{sub}_{ses}.tsv",sep="\t",header=None,index=False)
df_three_imagenet.to_csv(f"three_columns_{sub}_{ses}/custom_three_imagenet_{sub}_{ses}.tsv",sep="\t",header=None,index=False)
df_three_scene.to_csv(f"three_columns_{sub}_{ses}/custom_three_scenes_{sub}_{ses}.tsv",sep="\t",header=None,index=False)

```
